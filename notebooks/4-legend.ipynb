```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2 * x
y2 = x ** 2
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['line1'])
plt.show()
```

```python
plt.plot(x, y1, label = "linear line") # legend需要label才能展示
plt.plot(x, y2)
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, label = "linear line") # legend需要label才能展示
plt.plot(x, y2)
plt.legend(loc = "upper right") # 其他loc参数: upper right, upper left
plt.show()
```

```python
plt.plot(x, y1, label = "linear line") # legend需要label才能展示
plt.plot(x, y2, label = "line 2")
plt.legend(loc = 0, title = "legend title", shadow=True, ncol = 2, facecolor = "gray")
plt.show()
```

## Annotation

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.annotate("xxxx", xy = (1, 2), xytext= (2, 0), arrowprops = dict(arrowstyle="->"))
plt.legend(['line1', 'line2'])
plt.show()
```

## Text

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.text(1, 2, "cool")
plt.show()
```
