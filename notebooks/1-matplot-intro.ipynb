```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(-5, 5, 50) # 50个范围在(-5, 5)之间的等分点
y1 = 2 * x1
y2 = x1 ** 2 - 10
```

```python
plt.plot(x1, y1)
plt.show()
```

```python
plt.figure(figsize = (5, 5))
plt.plot(x1, y1)
plt.plot(x1, y2, color="green", linewidth=3.0, linestyle="--")
plt.savefig("./images/2-1.png")
```

```python
fig, axe = plt.subplots()
axe.plot(x1, y1)
plt.show()
```

```python
plt.subplot(2, 2, 1) # (nRows, nColumns, axes number to plot)
plt.plot(x1, y1)
plt.subplot(2, 2, 2)
plt.plot(x1, y2)
plt.show()
```

```python
fig = plt.figure()
plot = fig.add_subplot(121) 
plt.plot(x1, y1)
plot = fig.add_subplot(122)
plt.plot(x1, y2)
```
