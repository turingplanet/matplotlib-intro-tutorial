```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 50)
y1 = 2 * x
y2 = x ** 2 
```

## 基本设置: Title & Label

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.title("Title", color = "red")
plt.show()
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

## 坐标轴范围 Lim

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim((-2, 2))
plt.ylim((-5, 10))
plt.show()
```

```python
fig, axe = plt.subplots()
axe.plot(x, y1)
axe.plot(x, y2)
axe.set_xlim((-2, 2))
axe.set_ylim((-5, 10))
plt.show()
```

## 坐标轴记号和记号标签 Tick

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim((-2, 2))
plt.ylim((-5, 10))
plt.xticks(np.linspace(-2, 2, 3)) # only show -2 to 2
plt.show()
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xticks(np.linspace(-2, 2, 3))
plt.show()
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim((-2, 2))
plt.ylim((-2, 2))
plt.xticks([-2, 0, 2], ["x1", "x2", "x3"])
plt.yticks([-2, -1, 0, 1, 2], ["y1", "y2", "y3", "y4", "y5"])
plt.show()
```

```python
fig, axe = plt.subplots()
axe.plot(x, y1)
axe.plot(x, y2)
axe.set_xlim((-2, 2))
axe.set_ylim((-5, 10))
axe.set_xticks(np.linspace(-2, 2, 3))
plt.show()
```

```python
fig, axe = plt.subplots()
axe.plot(x, y1)
axe.plot(x, y2)
axe.set_xlim((-2, 2))
axe.set_ylim((-5, 10))
axe.set_xticks(np.linspace(-2, 2, 3))
axe.tick_params(direction = 'in', length = 5, width = 2, colors = 'b')
plt.show()
```

## 边框 Spines

```python
fig, axe = plt.subplots()
axe.plot(x, y1)
axe.plot(x, y2)
axe.set_xticks(np.linspace(-5, 5, 9))
axe.spines['left'].set_position(('axes', 0.5))
axe.spines['bottom'].set_position(('axes', 0.5))
plt.show()
```

```python
fig, axe = plt.subplots()
axe.plot(x, y1)
axe.plot(x, y2)
axe.set_xticks(np.linspace(-5, 5, 9))
axe.spines['left'].set_position(('axes', 0.5))
axe.spines['bottom'].set_position(('axes', 0.5))
axe.spines['top'].set_color('none')
axe.spines['right'].set_color('none')
plt.show()
```

## Colors

```python
fig, axe = plt.subplots()
y3 = np.tanh(x) + 0.5
y4 = np.sin(x) - 0.2
axe.plot(x, y1, color = (0.1, 0.6, 1.0), label = "rgb")
axe.plot(x, y2, color = "#6F4F3F", label = "hex")
axe.plot(x, y3, color = "r", label = "single letter")
axe.plot(x, y4, color = "green", label = "color name")
axe.legend()
plt.show()
```
